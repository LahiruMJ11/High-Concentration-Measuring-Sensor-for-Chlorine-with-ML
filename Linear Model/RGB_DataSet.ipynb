# Remove Outliers

```python
import pandas as pd

# Your data
X = df[['R', 'G', 'B']]
y = df['PPM']

# Combine features and target for outlier detection
data = pd.concat([X, y], axis=1)

# Compute IQR for each column
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Filter: Keep rows within 1.5 * IQR range
filtered_df = data[~((data < (Q1 - 1.5 * IQR)) |
                     (data > (Q3 + 1.5 * IQR))).any(axis=1)]

# Now separate back
X_clean = filtered_df[['R', 'G', 'B']]
y_clean = filtered_df['PPM']

x = X_clean.to_numpy()  # convert to numpy array
y = y_clean.to_numpy()  # convert to numpy array
```

# Compute Cost With Multiple Variables

```python
import numpy as np
```

```python
w_init = np.zeros(x_train.shape[1])  # initialize weights to zeros
b_init = 0  # initialize bias to zero
```

```python
def compute_cost(x, y, w, b):
    m,n = x.shape
    cost = 0 # initialize cost
    for i in range(m):
        f_wb_i = np.dot(x[i],w) + b
        cost += (f_wb_i - y[i]) ** 2
    cost = cost / (2 * m)  # compute average cost
    return cost
```

```python
cost = compute_cost(x_train, y_train, w_init, b_init)
print(f"Initial cost: {cost:.6f}")
```

# Compute Gradient  With Multiple Variables

```python
def compute_gradient(x,y,w,b):
    m,n = x.shape
    dj_dw = np.zeros(n)
    dj_db = 0
    for i in range(m):
        f_wb_i = np.dot(x[i],w) + b
        error = f_wb_i - y[i]
        for j in range(n):
            dj_dw[j] += error * x[i][j]  # accumulate gradient for each feature
        dj_db += error  # accumulate gradient for bias
    dj_dw = dj_dw / m
    dj_db = dj_db / m

    return dj_dw, dj_db
```

# Gradient Descent With Multiple Variables

```python
def gradient_descent(x,y,w_init,b_init,cost_func,gradient_func, alpha, num_iters):
    m,n = x.shape
    w = w_init.copy()
    b = b_init
    J_history = []  # to store cost history

    for i in range(num_iters):
        dj_dw, dj_db = gradient_func(x,y,w,b)
        w -= alpha * dj_dw  # update weights
        b -= alpha * dj_db  # update bias
        cost = cost_func(x,y,w,b)  # compute cost
        J_history.append(cost)  # store cost history

        if i % 100 == 0:  # print cost every 100 iterations
            print(f"Iteration {i}: Cost {cost:.6f}")

    return w, b, J_history
```

```python
iterations = 100000
# alpha = 1.2  # learning rate
alpha = 1
# run gradient descent
w_final, b_final, J_hist = gradient_descent(x_train, y_train, w_init, b_init,
                                                    compute_cost, compute_gradient,
                                                    alpha, iterations)
print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")
m,_ = x_train.shape
for i in range(m):
    print(f"prediction: {np.dot(x_train[i], w_final) + b_final:0.6f}, target value: {y_train[i]}")
```
